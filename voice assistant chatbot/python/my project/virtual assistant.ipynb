```python
from  neuralintents import GenericAssistant
import speech_recognition
import pyttsx3 as  tts
import sys
```

```python
recognizer = speech_recognition.Recognizer()

speaker = tts.init()
speaker.setProperty("rate", 150)
```

```python
todo_list = ["clean room", "record video", "present my project"]
```

```python
def note():
    global recognizer
    
    speaker.say("what do you want to note?")
    speaker.runAndWait()
    
    done = False
    
    while not done:
        try:
            
            with speech_recognition.Microphone()as mic:
                recognizer.adjust_for_ambient_noise(mic, duration = 0.3)
                audio = recognizer.listen(mic)
            
                note = recognizer.recognize_google(audio)
                note = note.lower()
            
                speaker.say("choose a filename!")
                speaker.runAndWait()
            
                recognizer.addjust_for_ambient_noise(mic, duration = 0.3)
                audio = recognizer.listen(mic)
            
                filename = recognizer.recognize_google(audio)
                filename = filename.lower()
            
            with open(filename, 'w') as f:
                f.write(note)
                done = True 
                
                speaker.say(f"I successfully created the note the note {filename}")
                speaker.runAndWait()
                
        except speech_recognition.UnknownValueError:
            recognizer = speech_recognition.Recognizer()
            speaker.say("i did not understand you! Please try again!")
            speaker.runAndWait()

```

```python
def add_todo():
    global recognizer
    
    
    speaker.say("what are you adding on your todo?")
    speaker.runAndWait()
    
    done  = False
    
    while not done:
        try:
            with speech_recognition.Microphone()as mic:
                recognizer.adjust_for_ambient_noise(mic, duration = 0.3)
                audio = recognizer.listen(mic)
                
                todo = recognizer.recognize_google(audio)
                todo = todo.lower()
                
                todo_list.append(todo)
                done = True
                
                speaker.say(f"i successfully added {todo}")
                speaker.runAndWait()
                
        except speech_recognition.UnknownValueError:
            recognizer = speech_recognition.Recognizer()
            speaker.say("i did not understand you! Please try again!")
            speaker.runAndWait()
            
        

            
            
```

```python
def show_todo():
    speaker.say("you have the following in your todo list")
    for todo in todo_list:
        speaker.say(todo)
    speaker.runAndWait()
```

```python
def hallo():
    speaker.say("Hello. How can i assist you")
    speaker.runAndwait()
```

```python
def quit():
    speaker.say("Good Bye")
    speaker.runAndWait()
    sys.exit(0)
```

```python
mapping = {
    "greeting": hallo,
    "create_note" : note,
    "add_todos": add_todo,
    "show_todos": show_todo,
    "exit": quit
}
```

```python
assistant = GenericAssistant("intents.json", intent_methods = mapping)
assistant.train_model()

while True:
    try:
        with speech_recognition.Microphone()as mic:
            recognizer.adjust_for_ambient_noise(mic, duration = 0.3)
            audio = recognizer.listen(mic)
            message = recognizer.recognize_google(audio)
            message = message.lower()
            
        assistant.request(message)
    except speech_recognition.UnknownValueError:
        recognizer = speech_recognition.Recognizer()
        
```
